NUMERICAL INTEGRATION

```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from scipy.integrate import quad
```

TRAPEZOIDAL METHOD

```python
def f(x): #define a function
  return m.exp(x)*m.cos(x) 
def trap(a,b,n):#a is lower limit b is upper limit n is no. of slices
 h=(b-a)/n #size of each slice
 Int=f(a)+f(b) # integral=h/2[f(a)+f(b)+(2*f(k))] 
 
 for i in range(1,n):
   k=a+i*h
   Int=Int + 2*f(k)
 Int=Int*h/2
 return Int
```

```python
trapval=trap(0,m.pi,100000)
print(trapval)
```

```python
#plotting deviation between trapezoidal and absolute value of integral taking scipy integral value as reference
n=[]
dev=[]
for i in range(1,51):
  n.append(i)
for i in n:
  error=abs(absval-trap(0,m.pi,i))
  dev.append(error)
print(dev)

plt.plot(n,dev)
```

SIMPSON 1/3 METHOD

```python
# simpson 1/3 method
def f(x): #defining function
  return m.exp(x)*m.cos(x)
def simp1(a,b,n): # a-lower limit,b-upper limit,n-no. of slices
  h=(b-a)/n #size of each slice
  Int=f(a)+f(b)
  #integral=h/3[f(a)+f(b)+2(sum of f(even terms))+4(sum of f(odd terms)))]     
  for i in range(1,n):
   if i%2==0:                   
    Int+=2*f(a+i*h)
   elif i%2!=0:                
    Int+=4*f(a+i*h)
  Int=h/3*Int
  return Int
```

```python
simpval=simp1(0,m.pi,100000)
print(simpval)
```

```python
#plotting deviation between simpson 1/3 and absolute value of integral taking scipy integral value as reference
n=[]
dev=[]
for i in range(1,51):
  n.append(i)
for i in n:
  error=abs(absval-simp1(0,m.pi,i))
  dev.append(error)
print(dev)

plt.plot(n,dev)
```

SIMPSON 3/8 METHOD

```python
def f(x):# define the function
  return m.exp(x)*m.cos(x)
def simp2(a,b,n):# a-upper limit b- lower limit n- no. of slices
  h=(b-a)/n  #size of each slice
  #integral=3h/8[f(0)+f(n)+3(f(1)+f(2)+f(4)+f(5)...f(n-1))+2(f(3)+f(6)+f(9)+...f(n-3))]
  sum= f(a) + f(b) 
  for i in range(1,n):
    if i%3==0:
     sum=sum+2*f(a+i*h)
    else:
      sum=sum+3*f(a+i*h)
  sum=3/8*h*sum
  return sum
simp2(0,m.pi,100000)
```

```python
#plotting deviation between simpson 3/8 and absolute value of integral taking scipy integral value as reference
n=[]
dev=[]
for i in range(1,51):
  n.append(i)
for i in n:
  error=abs(absval-simp2(0,m.pi,i))
  dev.append(error)
print(dev)

plt.plot(n,dev)
```
