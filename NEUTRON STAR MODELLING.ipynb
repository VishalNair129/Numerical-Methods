```python
import matplotlib.pyplot as plt 
import math as m
import numpy as np
```

```python
alpha=1
beta=0.7636
gamma=5/3
```

```python
def g(p,r):
    return beta*(r**2)*(p**(1/gamma))
```

```python
def f(p,m,r):
    return -(alpha*(p**(1/gamma))*m)/(r**2)
```

```python
p =0.01
m = 0.001
r = 1
rfinal =12 
delta_r = 0.001
N_slices = int(rfinal/delta_r)
rs = [r]
ps = [p]
ms = [m]
```

RK2

```python
for i in range(N_slices):
  k1 = delta_r*f(p,m,r)
  q1 = delta_r*g(p,r)
  k2 = delta_r*f(p+k1/2, m+q1/2, r+delta_r/2)
  q2 = delta_r*g(p+k1/2,r+delta_r/2)
  p += k2
  m += q2
  r += delta_r
  ps.append(p)
  ms.append(m)
  rs.append(r)
```

```python
plt.plot(rs,ps)
```

```python
plt.plot(rs,ms)
```

```python
for i in range(N_slices):
    k=delta_r*f(p,m,r)
    l=delta_r*g(p,r)
    k1=delta_r*f(p+k/2,m+l/2,r+delta_r/2)
    l1=delta_r*g(p+k/2,r+delta_r/2)
    k2=delta_r*f(p+k1/2,m+l1/2,r+delta_r/2)
    l2=delta_r*g(p+k1/2,r+delta_r/2)
    k3=delta_r*f(p+k2,m+l2,r+delta_r)
    l3=delta_r*g(p+k2,r+delta_r)
    p += 1/6*(k+2*k1+2*k2+k3)
    m += 1/6*(l+2*l1+2*l2+l3)
    r += delta_r
    ps.append(p)
    ms.append(m)
    rs.append(r)
    
```

```python
plt.plot(rs,ps)

```
